### Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title')["Book-Rating"].count().reset_index()
num_rating_df 
```

```python
num_rating_df.rename(columns={'Book-Rating':"num_rating"},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':"avg_rating"},inplace=True)
avg_rating_df
```

```python
popular_df = avg_rating_df.merge(num_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df["num_rating"]>=250]
popular_df
```

```python
popular_df = popular_df.sort_values('avg_rating',ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
popular_df
```

### Collaborative Filtering Based Recommender System

```python
x = filtered_rating = ratings_with_name.groupby('User-ID').count()["Book-Rating"] >200
x
```

```python
x[x]
```

```python
padhe_likhe_users = x[x].index # It will give you the user ID of padhe likhe users
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] > 50
famous_books = y[y].index # It will give you the names of the famouse books
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_rating
```

```python
pt = final_rating.pivot_table(index = 'Book-Title',columns = 'User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(pt)
similarity
```

```python
similarity.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    recommendations = sorted(list(enumerate(similarity[index])),key=lambda x:x[1],reverse=True)[1:6]

    for i in recommendations:
        print(pt.index[i[0]])
```

```python
recommend('1984')
```
